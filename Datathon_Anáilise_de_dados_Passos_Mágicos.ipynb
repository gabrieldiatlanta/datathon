```python
def indicadores(df):
    # Garantindo que as colunas existam no DataFrame
    colunas_desejadas = ["Nome", "INDE 2024", "IAN", "IDA", "IEG", "IAA", "IPS", "IPP", "IPV"]

    # Verificando se todas as colunas existem no DataFrame
    colunas_existentes = [col for col in colunas_desejadas if col in df.columns]

    # Criando uma cópia do dataset com as colunas necessárias
    df_filtrado = df[colunas_existentes].copy()

    return df_filtrado
```

```python
ordem_pedras = ["Quartzo", "Agata", "Ametista", "Topazio"]
def padronizar_pedra(pedra):
    if isinstance(pedra, str):
        return pedra.strip().lower().replace('á', 'a').capitalize()  # Remove espaços extras, padroniza acento
    return pedra

# Aplicando a função de padronização em todas as colunas de interesse
for coluna in ['Pedra 20', 'Pedra 21', 'Pedra 22', 'Pedra 23', 'Pedra 2024']:
    df_alunos_historico_pedra[coluna] = df_alunos_historico_pedra[coluna].apply(padronizar_pedra)



# Criando um dicionário para armazenar os dados por ano
dados_pedra = {
    '2020': df_alunos_historico_pedra['Pedra 20'].value_counts(),
    '2021': df_alunos_historico_pedra['Pedra 21'].value_counts(),
    '2022': df_alunos_historico_pedra['Pedra 22'].value_counts(),
    '2023': df_alunos_historico_pedra['Pedra 23'].value_counts(),
    '2024': df_alunos_historico_pedra['Pedra 2024'].value_counts()
}

# Criando um DataFrame consolidado
df_pedra_historico = pd.DataFrame(dados_pedra).fillna(0).astype(int)

df_pedra_historico = df_pedra_historico.reindex(ordem_pedras)

# Criando o gráfico de linhas
plt.figure(figsize=(10, 6))

for pedra in df_pedra_historico.index:
    plt.plot(df_pedra_historico.columns, df_pedra_historico.loc[pedra], marker='o', label=pedra)

# Configurações do gráfico
plt.title('Evolução das Pedras de 2020 a 2024')
plt.xlabel('Ano')
plt.ylabel('Número de Alunos')
plt.legend(title="Pedras")
plt.grid(True)
plt.xticks(rotation=45)

# Exibir o gráfico
plt.show()
```

```python
# Definição da ordem das pedras
ordem_pedras = {"Quartzo": 0, "Agata": 1, "Ametista": 2, "Topazio": 3}
```

```python
# Função para calcular o rendimento de cada aluno
def calcular_rendimento(row):
    # Determinar a pedra do ano de ingresso
    ano_ingresso = row["Ano ingresso"]
    colunas_pedra = ["Pedra 20", "Pedra 21", "Pedra 22", "Pedra 23"]

    # Encontrar a pedra correspondente ao ano de ingresso
    pedra_ingresso = None
    for coluna in colunas_pedra:
        if pd.notna(row[coluna]):
            pedra_ingresso = row[coluna]
            break

    # Se não encontrou nenhuma pedra, assume que o aluno se manteve
    if pedra_ingresso is None:
        return "M"

    pedra_atual = row["Pedra 2024"]

    # Comparar os índices das pedras
    if ordem_pedras[pedra_ingresso] < ordem_pedras[pedra_atual]:
        return "P"
    elif ordem_pedras[pedra_ingresso] > ordem_pedras[pedra_atual]:
        return "R"
    else:
        return "M"

```

```python
# Aplicar a função calcular_rendimento ao DataFrame
df_aluno.loc[:, "rendimento"] = df_aluno.apply(calcular_rendimento, axis=1)
```

```python
# Mapeamento de valores
mapeamento_pedra = {
    "Quartzo": 1,
    "Agata": 2,
    "Ametista": 3,
    "Topazio": 4
}
```

```python
# Lista das colunas de Pedra
colunas_pedra = ["Pedra 20", "Pedra 21", "Pedra 22", "Pedra 23", "Pedra 2024"]
```

```python
# Criar uma lista para armazenar os novos registros
dados_transformados = []

# Iterar sobre cada linha do dataset original
for _, row in df_aluno.iterrows():
    for col in colunas_pedra:
        ano = col.split()[-1]
        ano = f"20{ano}" if len(ano) == 2 else ano  # Ajustar apenas anos com 2 caracteres
        valor_pedra = row[col] if pd.notna(row[col]) else None  # Obter o valor da pedra

        if valor_pedra in mapeamento_pedra:
            dados_transformados.append({
                "RA": row["RA"],
                "Pedra": valor_pedra,
                "Ano": ano,
                "Valor": mapeamento_pedra[valor_pedra]
            })
```

```python
# Criar o novo DataFrame
df_pedras_por_aluno = pd.DataFrame(dados_transformados)
```
